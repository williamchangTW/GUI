```python
import sys
import os
from PyQt5.QtWidgets import *
from PyQt5.QtCore import *
from PyQt5.QtGui import *

class MainForm(QMainWindow):
    def __init__(self):
        super().__init__()
        self.title = "Main Control"
        self.cwd = os.getcwd() # 获取当前程序文件位置
        self.width = 700
        self.height = 500
        self.top = 100
        self.left = 100
        self.InitUI()
        
    def InitUI(self):
        # Initial Main Frame
        self.setWindowTitle(self.title)
        self.setGeometry(self.top, self.left, self.width, self.height)
        # Select Model File
        self.btn_chooseFile = QPushButton(self)
        self.btn_chooseFile.move(100, 100)
        self.btn_chooseFile.setObjectName("btn_chooseFile")  
        self.btn_chooseFile.setText("Input CSV File")
        self.btn_chooseFile.setGeometry(250, 100, 400, 30)

        # Show devices connected
        self.btn_chooseMutiFile = QPushButton(self)
        self.btn_chooseMutiFile.move(100, 200)
        self.btn_chooseMutiFile.setObjectName("btn_chooseMutiFile")  
        self.btn_chooseMutiFile.setText("Input Model File")
        self.btn_chooseMutiFile.setGeometry(250, 200, 400, 30)
        
        # new window to switch
        self.btn_newwindow = QPushButton(self)
        self.btn_newwindow.move(100, 300)
        self.btn_newwindow.setObjectName("btn_chooseInputFile")
        self.btn_newwindow.setText("New Window")
        self.btn_newwindow.setGeometry(250, 300, 400, 30)
        
        self.show()


        # 设置信号
        #self.btn_newwindow.clicked.connect(self.slot_btn_newwindow)
        self.btn_chooseFile.clicked.connect(self.slot_btn_chooseInputFile)
        self.btn_chooseMutiFile.clicked.connect(self.slot_btn_chooseModelFile)
        #self.btn_saveFile.clicked.connect(self.slot_btn_saveFile)

    def slot_btn_chooseInputFile(self):
        fileName_choose, filetype = QFileDialog.getOpenFileName(self,  
                                    "Select",  
                                    self.cwd, # 起始路径 
                                    "CSV Files (*.csv)")   # 设置文件扩展名过滤,用双分号间隔

        if fileName_choose == "":
            print("\nCancel")
            return

        print("\nFile Name:")
        print(fileName_choose)
        print("File Filter: ",filetype)
        
    def slot_btn_chooseModelFile(self):
        fileName_choose, filetype = QFileDialog.getOpenFileName(self,  
                                    "Select",  
                                    self.cwd, # 起始路径 
                                    "Python File(*.py)")   # 设置文件扩展名过滤,用双分号间隔

        if fileName_choose == "":
            print("\nCancel")
            return

        print("\nFile Name:")
        print(fileName_choose)
        print("File Filter: ",filetype)
        
        
if __name__ == "__main__":
    app = QApplication(sys.argv)
    mainForm = MainForm()
    sys.exit(app.exec_())
```
